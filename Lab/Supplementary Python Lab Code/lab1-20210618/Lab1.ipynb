# Random Numbers

There are many ways to generate pseudo-random values. We shall use a function that generates pseudo-random Normally distributed values. These are numbers that, when plotted in sufficient numbers would show a bell-shaped histogram. The function ```np.random.normal(size=50)``` will generate 50 normally distributed (random) values that have a mean of zero and a standard deviation of one (since these are the defaults).
 The second line (below) will explicitly specify a mean of 50 and a standard deviation of 0.1.


````python
x = np.random.normal(size=50)
y = x + np.random.normal(50, .1, 50)
````

This has created two columns that have a carefully designed relationship to each other - due to the second column referencing the first column. How to we confirm this relationship? Let’s try correlation.

What is the correlation between x and y?


````python
np.corrcoef(x, y)
````

While x and y are columns (of the same length: 50), their correlation is a scalar (i.e. a column of length 1)

Pseudo-random generators make use of a “seed”" value from which a sequence of values will stream.


````python
np.random.seed(1303)
np.random.normal(size=50)
````

Changing the seed, will generate a different sequence

````python
np.random.seed(3)
y = np.random.normal(size=50)
````

We can generate summary statistics for this column of numbers:

````python
np.mean(y)
````

````python
np.var(y)
````

````python
np.sqrt(np.var(y))
````

````python
np.std(y)
````

Suppose we pretend to have two columns that are meant to be very similar

````python
y = np.asarray([1, 2, 2.5, 3, 4]) # measured values
hat_y = np.asarray([1.1, 1.9, 2, 2.8, 4.2]) # predicted values
````

You can also calculate a Mean Squared Error (MSE) as follows

````python
mse = (np.square(y - hat_y)).mean()
mse
````

... or the long way:

````python
mse = (1 / len(y)) * sum((y - hat_y) ** 2)
mse
````

# Complex graphics

Let’s create a fancy contour plot using this data (as x & y) and a function that represents the z axis


````python
x = np.linspace(-np.pi, np.pi, 50)
y = x

f = np.frompyfunc(lambda i, j: np.cos(j) / (1 + i ** 2), 2, 1)
z = f.outer(x, y)

contour = plt.contour(x, y, z, colors="black")
plt.clabel(contour, inline=True, fontsize=8)
````

Here we used an in-situ function that generated the 3rd dimension from x and y. The outer function makes a matrix spanning x and y in which the values are the z values (calculated by the function)

We can add more contours

````python
contour = plt.contour(x, y, z, 45, colors="black")
plt.clabel(contour, inline=True, fontsize=8)
````

We can perform matrix operations on this matrix.

````python
za = (z - z.transpose()) / 2
contour = plt.contour(x, y, za, 15, colors="black")
plt.clabel(contour, inline=True, fontsize=8)
````

The use of ```plt.contourf``` with ```cmap="hot"``` replaces contours with colour gradients

````python
za = (z - z.transpose()) / 2
plt.contourf(x, y, za, 15, cmap="hot")
````

# Indexing Data

Indexing is the means of locating sets of positions within matrices and columns.


````python
a = np.reshape(np.arange(1, 17), (4, 4), "F")
a
````

Suppose we want the value at row 2, column 3

````python
a[1, 2]
````

````python
a[(0, 2),][:, (1, 3)]
````

We can use sequences in place of explicit columns:

````python
a[0:3, 1:4]
````

We can leave an aspect undefined. The default is everything in that dimension. The following will (first of all) return all values of rows 1 to 2

````python
a[0:2, ]
````

````python
a[:, 0:2]
````

````python
a[0, ]
````

The ```.shape``` attributes returns the dimensions of a matrix

````python
a[0:3, 1:4].shape
````
